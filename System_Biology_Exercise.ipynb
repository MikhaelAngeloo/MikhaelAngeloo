```python
import random 
import string

# Random String generator V1

def randomword1(length):
  return ''.join(random.choice(string.ascii_lowercase) for i in range(length))

# Random String Generator v2

def randomword2(length):
  x = ""
  for i in range(length):
    x = x + random.choice(string.ascii_lowercase)
  return x

# returns a 10 characters long string by calling function v1
randomString1 = randomword1(10)
# returns a 10 characters long string by calling function v2
randomString2 = randomword2(15)

print(randomString1)

```
